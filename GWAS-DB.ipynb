# Analyze the function **load_gwas_cat()** in *LoadData.py*

```python
import os
import pandas as pd
```

## load the cat_stud data file

```python
Cat_Stud = pd.read_csv(os.path.abspath(
                           os.path.join('__file__',
                                        '..',
                                        'data',
                                        'Catalogue',
                                        'Raw',
                                        'Cat_Stud.tsv')),
                           header=0, sep='\t', encoding='utf-8',
                           index_col=False)
Cat_Stud.head()
```

```python
# fill in the missing value with NA
Cat_Stud.fillna('N/A', inplace=True)
Cat_Stud.head()
```

## load the ancenstry data *Cat_Anc*

```python
Cat_Anc = pd.read_csv(os.path.abspath(
                          os.path.join('__file__', '..',
                                       'data',
                                       'Catalogue',
                                       'Raw',
                                       'Cat_Anc.tsv')),
                          header=0, sep='\t', encoding='utf-8',
                          index_col=False)

Cat_Anc.rename(columns={'BROAD ANCESTRAL CATEGORY': 'BROAD ANCESTRAL',
                            'NUMBER OF INDIVDUALS': 'N'}, inplace=True)
Cat_Anc.head()
# if an article contains more than one ancestry, then the data will breakdown the ancenstries and put them into different rows.
```

```python
# drop those with null ancenstry articles
Cat_Anc = Cat_Anc[~Cat_Anc['BROAD ANCESTRAL'].isnull()]
Cat_Anc.head()
```

```python
#Replace occurrences of pattern in the the column names with some other string. Here is used to correctify the spelling
Cat_Anc.columns = Cat_Anc.columns.str.replace('ACCCESSION', 'ACCESSION')
Cat_Anc.head()
```

## Collapse by article, and obtain the sample size information for each article.
Remark: Currently this sum of sample size for each article is the total of replication sample and initial sample.

```python
Cat_Anc_byN = Cat_Anc[['STUDY ACCESSION', 'N',
                         'DATE']].groupby(by='STUDY ACCESSION').sum()
Cat_Anc_byN = Cat_Anc_byN.reset_index()
Cat_Anc_byN.head()

# get the initial sample size fo each article
# Cat_Anc_byN = Cat_Anc[Cat_Anc["STAGE"] == "initial"][["STUDY ACCESSION", "N", "DATE"]].groupby("STUDY ACCESSION").sum()
```

```python
# merge the ancestral summary back with the gwas article info.
Cat_Anc_byN = pd.merge(Cat_Anc_byN, Cat_Stud[[
        'STUDY ACCESSION', 'DATE']], how='left', on='STUDY ACCESSION')
Cat_Anc_byN.head()
```

## Begin to clean the ancestral categories based on a per-specified ancestral cleaner

#### Load in the ancestral cleaner

```python
cleaner_broad = pd.read_csv(os.path.abspath(
        os.path.join('__file__',
                     '..',
                     'data',
                     'Support',
                     'dict_replacer_broad.tsv')),
        sep='\t', header=0, index_col=False)
cleaner_broad.head()
```

#### merge the **ancenstral cleaner** to the ancenstry data

```python
Cat_Anc = pd.merge(Cat_Anc, cleaner_broad, how='left',
                       on='BROAD ANCESTRAL')
Cat_Anc.head()
# REMARK: the column "Broader" is the merged info from the ancestral cleaner.
```

```python
Cat_Anc['Dates'] = [pd.to_datetime(d) for d in Cat_Anc['DATE']]
# creates a new date column "Dates", but why?
#print(type(Cat_Anc.loc[0,'DATE']))  The column "DATE" 's element is a str
#print(type(Cat_Anc.loc[0, "Dates"])) The column "Dates"'s element is a time type
Cat_Anc.head()
```

```python
print(type(Cat_Anc.loc[0,"N"]))
Cat_Anc['N'] = pd.to_numeric(Cat_Anc['N'], errors='coerce')
Cat_Anc.head()
```

```python
## keep the non-missing sample size articiles
Cat_Anc = Cat_Anc[Cat_Anc['N'].notnull()]

## cast the sample size column into integer, previous is float
Cat_Anc['N'] = Cat_Anc['N'].astype(int)

## sort the data by dates, from the earliest to latest
Cat_Anc = Cat_Anc.sort_values(by='Dates', ascending = True)

Cat_Anc.head()
```

```python
Cat_Anc['Broader']
## Hard coding: replace a typical ancestry category, there pandas.series.str.replace is used.
Cat_Anc['Broader'] = Cat_Anc['Broader'].str.replace(
        'African American or Afro-Caribbean', 'African Am./Caribbean')

Cat_Anc['Broader'] = Cat_Anc['Broader'].str.replace(
        'Hispanic or Latin American', 'Hispanic/Latin American')

Cat_Anc = Cat_Anc.sort_values(by=["PUBMEDID", 'Dates'], ascending = True)
Cat_Anc.head()
```

## Load the full GWAS article info dataset, which includes all the information, such as p-value, SNP, for each GWAS article.

```python
Cat_Full = pd.read_csv(os.path.abspath(os.path.join('__file__',
                                                        '..',
                                                        'data',
                                                        'Catalogue',
                                                        'Raw',
                                                        'Cat_Full.tsv')),
                           header=0, sep='\t', encoding='utf-8',
                           index_col=False, low_memory=False)
Cat_Full.head()
```

## Save the merged ancestry data set into csv file.

```python
Cat_Anc.to_csv(os.path.abspath(
        os.path.join('__file__',
                     '..',
                     'data',
                     'Catalogue',
                     'Synthetic',
                     'Cat_Anc_withBroader.tsv')),
                   sep='\t', index=False)
```
